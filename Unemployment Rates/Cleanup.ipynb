```python
# All variables that can be discarded
cols_drop = ['men_16_17_rate',
       'women_16_17_rate','men_18_19_rate', 'women_18_19_rate', 'men_16_24_rate',
       'women_16_24_rate','men_25plus_rate', 'women_25plus_rate', 'men_25_54_rate', 'women_25_54_rate']

# Dropping variables we want to discard
data.drop(columns=cols_drop, inplace = True)
```

The only variable with missing values is women aged 55+

```python
# Checking for the dates where there are no entries for women at 55+
missing_dates = data[data['women_55plus_rate'].isna()]['date']
print("Dates with missing values:")
print(missing_dates)
```

With a total of 907 entries for each age range. There were 355 non null values and 552 entries with missing values for women 55+, which total 907.
We can conclude from this that there were not any entries recorded for women at 55+ between the years 1948 and 1993.
This column will not be removed due to its importance in later years.
One theory could be that women in this age group did not work, or the amount of workforce for that age group was very low that it did not get recorded.

```python
data.describe().T
```

Using the describe method we can gain an understanding that younger people tend to have a higher unemployment rate, the rate decreases as they frown in age and at the age of 35 and older the unemployment rate seems to be constant.
As a mean it seems that men have a higher unemployment rate, but looking at the Max values women tend to have a higher unemployment rate. We could conclude from this that factor that cause increased unemployment rates affect women more than men.

Due to the comparison between men_45_54_rate and men_55plus_rate is also close while having no missing values, and the comparison between women_45_54_rate and women_55plus_rate is very close with women_55plus_rate having missing values.
We can fill in for the missing values for women_55plus_rate according to women_45_54_rate without disrupting the validity of the data

```python
#  Function for filling missing values for women 55+ 
def fill_NA(row):
    if pd.isna(row['women_55plus_rate']):
        # reduction of 0.13 to each value is used as it is the difference between the mean values of women 45-54 and women 55+
        return row['women_45_54_rate'] - 0.13
    else:
        return row['women_55plus_rate']

# Applying the function
data['women_55plus_rate'] = data.apply(fill_NA, axis=1)
```

```python
# There is a slight change in the values for women_55plus_rate but not enough to cause a disruption
data.describe().T
```
